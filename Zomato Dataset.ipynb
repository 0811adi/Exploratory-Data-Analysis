```python
# changing column name
rating = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().\
rename (columns = {0:"Rating Count"})
```

```python
rating
```

```python
import matplotlib
matplotlib.rcParams["figure.figsize"] = (12,6)  ## for changing the size of plot
sns.barplot(x = "Aggregate rating", y = "Rating Count",data = rating)
```

```python
# to get the colour
sns.barplot(x = "Aggregate rating", y = "Rating Count",hue = "Rating color", data = rating)
```

```python
## for mapping the colour
sns.barplot(x = "Aggregate rating", y = "Rating Count",hue = "Rating color", \
            palette = ['blue',"red","orange","yellow","green","green"] ,data = rating)
```

```python
# Count plot  ## it is frequency count
sns.countplot(x="Rating color",data = rating, palette = ['blue',"red","orange","yellow","green","green"])
```
